```python
import pandas as pd
import numpy as np

tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", attrs={"class": "wikitable collapsible collapsed"})
# print(tmp)

# print(tmp[3])
df = tmp[3]
```

 Вивести перші рядки таблиці за допомогою методу head

```python
print(df.head())
```

 Замініть у таблиці значення "—" на значення NaN

```python
df.replace('—', np.NaN, inplace=True)
print(df)
```

 Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
print(df.dtypes)
```

 Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
for column, dtype in zip(df.columns.tolist(), df.dtypes):
  if dtype == 'object' and column != 'Регіон':
    df[column] = df[column].astype(float)
  elif dtype == 'int64':
    df[column] = df[column].astype(float)
# print(df)
df.dtypes
```

 Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python

print(df.isnull().sum())
```

 Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df.drop(index=df.index[-1], inplace=True)
df
```

 Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python

df = df.fillna({i: df[i].mean() for i in df if i != 'Регіон'})

df
```

```python
import matplotlib.pyplot as plt
y1 = df['2019']
y2 = df['2014']
x = df['Регіон']
plt.bar(x, y2, label="2014")
plt.bar(x, y1, label="2019")
plt.xlabel('Регіони', fontsize='small', color='red')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
x = df.sum()
y = x.index.tolist()[1:]
x = x[1:]
x= x.tolist()

plt.plot(y, x, linestyle="--", color="red")
plt.title('Народжуванність в Україні', fontsize=15)
```

```python
x = df.columns.tolist()[1:]
y1 = df.loc[9].tolist()[1:]
y2 = df.loc[6].tolist()[1:]
y3 = df.loc[19].tolist()[1:]

fig, axs = plt.subplots()
axs.plot(x, y1, label='Київська')
axs.plot(x, y2, label='Закарпатська')
axs.plot(x, y3, label='Харківська')
plt.title('по регіонам', fontsize=15)
plt.legend()
plt.show()
```

 Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
avg_br = df['2019'].mean()

above_avg_br = df[df['2019'] > avg_br][['Регіон', '2019']]

print(above_avg_br)
```

 У якому регіоні була найвища народжуваність у 2014 році?

```python
max_br = df['2014'].max()
eq_avg_br = df[df['2014'] == max_br][['Регіон', '2014']]
print(eq_avg_br)
```
