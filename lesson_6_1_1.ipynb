```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import seaborn as sns
```

```python
datasets = [
    datasets.make_circles(n_samples=1000, factor=0.5, noise=0.05),
    datasets.make_moons(n_samples=1000, noise=0.05),
    datasets.make_blobs(n_samples=1000, random_state=8),
]
```

```python
plt.figure(figsize=(14, 5))
for i, data in enumerate(datasets):
    X, _ = data
    plt.subplot(1, 3, i + 1)
    plt.scatter(X[:, 0], X[:, 1])
```

```python
plt.figure(figsize=(14, 14))
for i, data in enumerate(datasets):
    X, _ = data
    X_norm = StandardScaler().fit_transform(X)

    plt.subplot(3, 4, 4 * i + 1)
    plt.scatter(X_norm[:, 0], X_norm[:, 1])

    kmeans = KMeans(n_clusters=2).fit(X_norm)
    plt.subplot(3, 4, 4 * i + 2)
    plt.scatter(X_norm[:, 0], X_norm[:, 1], c=kmeans.labels_)

    agg_cl = AgglomerativeClustering(n_clusters=None, distance_threshold=0.5, linkage="average").fit(X_norm)
    plt.subplot(3, 4, 4 * i + 3)
    plt.scatter(X_norm[:, 0], X_norm[:, 1], c=agg_cl.labels_)

    dbscan_cl = DBSCAN(eps=0.3).fit(X_norm)
    plt.subplot(3, 4, 4 * i + 4)
    plt.scatter(X_norm[:, 0], X_norm[:, 1], c=dbscan_cl.labels_)
```
