# Пошук та сортування даних

```python
max_rating = df['user_rating'].max()
print(f'Який рейтинг у датасеті найвищий? Відповідь: {max_rating}')

number_with_max_r = (df['user_rating'] == max_rating).sum()
print(f'Скільки книг мають такий рейтинг? Відповідь: {number_with_max_r}')

book_max_reviews = df[df['reviews'] == df['reviews'].max()]['name'].tolist()[0]
print(f'Яка книга має найбільше відгуків? Відповідь: {book_max_reviews}')

top50_2015 = df[df['year'] == 2015].nlargest(50, 'reviews')
most_expensive_top50_2015 = top50_2015[top50_2015['price'] == top50_2015['price'].max()]['name'].tolist()[0]
print(f'З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)? Відповідь: {most_expensive_top50_2015} ')

q_fiction_2010 = df[(df['genre'] == 'Fiction') & (df['year'] == 2010)].shape[0]
print(f'Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)? Відповідь: {q_fiction_2010}')

books_rat_year= df[(df['year'].isin([2010, 2011])) & (df['user_rating'] == 4.9)].shape[0]
print(f'Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? Відповідь: {books_rat_year}')

year_prise = df[(df['year'] == 2015) & (df['price'] < 8)].sort_values('price')
print(f'Яка книга остання у відсортованому списку? Відповідь:\n{year_prise}')
```

# Агрегування даних та з'єднання таблиць

```python
grouped = df.groupby('genre')
agg = grouped.agg({'price': ['min', 'max']})
agg
```

Максимальна ціна для жанру Fiction: 82

Мінімальна ціна для жанру Fiction: 0

Максимальна ціна для жанру Non Fiction: 105

Мінімальна ціна для жанру Non Fiction: 0


```python
grouped_authors = df.groupby('author')
agg_authors = grouped_authors.agg({'name': 'count'})
agg_authors.columns = ['books_count']
print(agg_authors.sort_values('books_count'))
```

Якої розмірності вийшла таблиця? Відповідь: [248 rows x 1 columns]

Який автор має найбільше книг? Відповідь: Jeff Kinney

Скільки книг цього автора? Відповідь: 12

```python
agg_authors_rating = grouped_authors.agg({'user_rating': 'mean'})
agg_authors_rating.columns = ['avg_user_rating']
print(agg_authors_rating.sort_values('avg_user_rating'))
```

У якого автора середній рейтинг мінімальний? Відповідь: 3.9

Який у цього автора середній рейтинг? Відповідь: 4.9

```python
b_r_authors = pd.concat([agg_authors, agg_authors_rating], axis=1)
print(b_r_authors.sort_values(['books_count', 'avg_user_rating']))
```

Який автор перший у списку? Відповідь: Muriel Barbery
